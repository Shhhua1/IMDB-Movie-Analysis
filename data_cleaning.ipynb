```python
import pandas as pd
import numpy as np
```

```python
# Loading in the dataset and doing an initial check
df = pd.read_csv('imdb_top_1000.csv')

df.head()
df.shape
df.columns
df.info()
df.describe()
df.sample(n=5, random_state=1)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Dropping the Poster URL column
df.drop(columns=['Poster_Link'], inplace=True)
df.head()
```

```python
# Making overview all lowercase
df['Overview'] = df['Overview'].str.lower()
```

```python
df.head()
```

```python
# removing special characters from the overview
df['Overview'] = df['Overview'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df.head()
```

```python
df.info()
```

```python
df.to_csv('imdb_cleaned_data.csv', index=False)
```
